```python
# Importing libraries

import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
sample_input = np.loadtxt("./data/A5_q2.csv", delimiter=',')
X_train = sample_input[:,0]
Y_train = sample_input[:,1]
```

```python
def simple_lr(X, Y):
    X_bar = np.mean(X)
    Y_bar = np.mean(Y)

    X_diff = X - X_bar
    Y_diff = Y - Y_bar

    beta_1 = np.dot(X_diff, Y_diff) / np.dot(X_diff, X_diff)
    beta_0 = Y_bar - (beta_1 * X_bar)

    Y_pred = beta_0 + beta_1 * X

    sse = np.sum(np.square(Y_pred - Y))
    mape = (100*np.sum(np.divide(np.abs(Y - Y_pred), Y)))/len(Y)
    
    return Y_pred, beta_0, beta_1, sse, mape
```

```python
def plot(X, Y, Y_pred, title):
    plt.plot(X, Y, label = 'Y_original')
    plt.plot(X, Y_pred, label = 'Y_pred')
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.legend(loc = 'upper left')
    plt.title(title)
    plt.show()
```

```python
# Question 2(a)

Y_pred, beta_0, beta_1, sse, mape = simple_lr(X_train, Y_train)

print("beta_0: " + str(beta_0))
print("beta_1: " + str(beta_1))
print("SSE: " + str(sse))
print("MAPE: " + str(mape))

plot(X_train, Y_train, Y_pred, 'Q2_a')
```

```python
# Question 2(b)
X = X_train
X = (1/(1-X/100))

Y_pred, beta_0, beta_1, sse, mape = simple_lr(X, Y_train)

print("beta_0: " + str(beta_0))
print("beta_1: " + str(beta_1))
print("SSE: " + str(sse))
print("MAPE: " + str(mape))

plot(X_train, Y_train, Y_pred, 'Q2_b')
```

```python
# Question 2(c)
X = X_train
X = (1/(1-((X/100)**2)))

Y_pred, beta_0, beta_1, sse, mape = simple_lr(X, Y_train)

print("beta_0: " + str(beta_0))
print("beta_1: " + str(beta_1))
print("SSE: " + str(sse))
print("MAPE: " + str(mape))

plot(X_train, Y_train, Y_pred, 'Q2_c')
```
